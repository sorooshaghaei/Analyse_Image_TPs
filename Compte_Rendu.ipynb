### TP4: Connected Operations and Segmentation (Session 4)

- **Status:** Completed.  
- **Objective:** Analyze pixel connectivity to label objects and segment the image according to homogeneity criteria.  
- **Implemented algorithms:**
  1. Connected Component Labeling (DFS): Depth-first search to identify distinct objects (4-connectivity).  

  2. Area Filter: Removal of components whose size (number of pixels) is below a threshold (noise cleaning).  

  3. Two-Pass Labeling: Optimized algorithm using a Union-Find structure to manage label equivalences.  

  4. Otsu’s Method: Automatic computation of the optimal threshold maximizing inter-class variance.  

  5. Region Growing: Segmentation by propagation from a seed according to a homogeneity criterion.  
- **Code:** Available in the `TP4/` folder.

### TP5: Morphology (Session 5)

- **Status:** Completed.  
- **Objective:** Modify the shape of objects in the image using set theory (structuring elements).  
- **Implemented operators:**
  1. Dilation: Expansion of bright regions (local max).  

  2. Erosion: Reduction of bright regions (local min).  

  3. Opening: Erosion followed by dilation (removal of white noise/small elements).  

  4. Closing: Dilation followed by erosion (filling dark holes).  

  5. Morphological Gradient: Difference between dilation and erosion to extract contours.  
- **Code:** Available in the `TP5/` folder.
